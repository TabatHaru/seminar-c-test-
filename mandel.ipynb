```python
import matplotlib.pyplot as plt
import numpy as np

# データの生成
x = np.linspace(-2, 2, 100)
y = np.exp(x)

# グラフの描画
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$e^x$')
plt.title('Exponential Function $y = e^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='grey', lw=0.8)
plt.axvline(0, color='grey', lw=0.8)
plt.legend()

plt.show()
```
